```python
from math import sin, cos, exp

#configure params here
a=1
b=1
step=1

th=0
numsteps = 10

xs = [a*cos(th)*exp(b*th)-1 for th in range(0, numsteps*step, step)]
ys = [a*sin(th)*exp(b*th) for th in range(0, numsteps*step, step)]

print(xs)

import matplotlib.pyplot as plt

plt.plot(xs, ys)
plt.show()
```
